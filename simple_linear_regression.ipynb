```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```
df=pd.read_csv("dataset.csv")
real_y=df["Salary"]
real_x=df["YearsExperience"]
real_x=real_x.to_numpy()
real_x=real_x.reshape(-1,1)
real_y=real_y.to_numpy()
```

```
lin=LinearRegression()
training_x,testing_x,training_y,testing_y=train_test_split(real_x,real_y)
```

```
lin.fit(training_x,training_y)
pred_y=lin.predict(testing_x)
```

```
plt.scatter(testing_x,testing_y,label="testing curve")
plt.plot(testing_x,pred_y,label="predicted curve")
plt.title("Linear Regression Curve")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
